```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Data for Numeric Columns Only
df.describe()
```

```python
# Data for Non-Numeric Columns
df.describe(include='all')
```

```python
# Pinpointing Null Values
df.isnull().sum()
```

```python
# Filling data with medians of their respective category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
# Re-checking Null values
df.isnull().sum()
```

```python
# Replacing column names to names with snake casing
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Creating Column of age groups - age_group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# Creating Purchase frequency days column to convert string to numeric, for easier analysis

df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping) 
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
# Checking whether both columns are equal i.e., if discounts are only available when promo codes are used

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# We can remove promo code, as discounts are only offered via promo codes
df = df.drop('promo_code_used', axis = 1)
```
